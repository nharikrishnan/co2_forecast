## Forecasting CO2 levels using Holt winters

<b><font size = 4>Objective</b>
* To analyze and forecast the atmospheric CO2 using holtwinters

```python
# Loading the required modules
import pandas as pd
import numpy as np
from statsmodels.tsa.api import ExponentialSmoothing
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm

import matplotlib.pyplot as plt
```

### Loding the dataset from statsmodels

```python
# Loading the dataset
co2 = sm.datasets.co2.load_pandas()
df_co2 = co2.data
print(df_co2.head())
```

```python
plt.figure(figsize = (16,8))
plt.plot(df_co2, label = 'atmospheric CO2')
plt.xlabel('year')
plt.ylabel('CO2 level')
plt.legend(loc = 'best')
plt.show()
```

<p> <font size = 3> Clearly there is an upward trend and seasonality

```python
#Checking for missing values 
print(df_co2.isna().sum())
```

```python
#treating na values
#replacing the na value with previour terms value
df_co2.fillna(df_co2.bfill(), inplace = True)
```

```python
#splitting the data into Train and Test datasets
df_co2.reset_index(inplace = True)
df_co2.columns = ['date', 'co2']
df_co2['date'] = pd.to_datetime(df_co2['date'])
df_co2.min(), df_co2.max()
```

<p><font size=3> For testing, the last three years data is used.

```python
# splitting the data trainning and testing
train = df_co2[df_co2['date'] <'1998-01-01']

#last 3 years data for testing
test = df_co2[df_co2['date']>='1998-01-01']
train.set_index('date', inplace=True)
test.set_index('date', inplace=True)
```

### Analyzing the data 

```python
# seasonal decompose- visualizing the trend, seasonal, and residual component
ts_decompose = seasonal_decompose(train, freq = 52)
trend = ts_decompose.trend
seasonal = ts_decompose.seasonal
residual = ts_decompose.resid

#training data
plt.figure(figsize = (30,20))
plt.subplot(411)
plt.plot(train, label = 'Original')
plt.legend(loc = 'best')

#Trend Component
plt.subplot(412)
plt.plot(trend, label ='Trend')
plt.legend(loc = 'best')

#Seasonal Component
plt.subplot(413)
plt.plot(seasonal, label = 'Seasonal')
plt.legend(loc = 'best')

#Residual Component
plt.subplot(414)
plt.plot(residual, label = 'residual')
plt.legend(loc = 'best')

plt.tight_layout()
plt.show()
```

### Forecasting using holt-winters

<p> <font size=3> Holt-winter Exponential smoothing can be used for forecasting time series that has both trend and seasonal components. There are 2 variations of holt-winters, one is the Additive method and the other one is the Multiplicative method.</font></p>
<p> <font size=3>Here, Mulitplicative method is used.

```python
fit_forecast = ExponentialSmoothing(np.asarray(train['co2']), seasonal_periods=52, seasonal = 'mul', trend = 'mul').fit()
y_pred = test.copy()

#forecast
y_pred['holt_winter'] = fit_forecast.forecast(len(test['co2']))

#plotting
plt.figure(figsize = (15, 6))
plt.plot(train['co2'], label='train')
plt.plot(test['co2'], label='Actual')
plt.plot(y_pred['holt_winter'], label = 'Forecast')

plt.legend(loc = 'best')

plt.show()
```

```python
# checking the root mean squared error
from sklearn.metrics import mean_squared_error
error = np.sqrt(mean_squared_error(test['co2'], y_pred['holt_winter']))
print(error)
```

```python
train['co2'].std()
```

```python
#plotting the difference in forecast
diff = test['co2'] - y_pred['holt_winter']
plt.plot(diff)
```

### Forecasting using ARIMA

```python
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.stattools import adfuller
```

```python
# Checking for stationarity
train_log = np.log(train['co2'])
test_log = np.log(test['co2'])
```

```python
# Reducing the scale
plt.figure(figsize = (15, 6))
plt.plot(train_log)
```

```python
def test_stationarity(ts):
    rol_mean = ts.rolling(52).mean()
    rol_std = ts.rolling(52).std()
    orig = plt.plot(ts, color='blue', label = 'Data')
    mean = plt.plot(rol_mean, color = 'red', label = 'rolling mean')
    #std_dev = plt.plot(rol_std, color = 'black', label ='rolling std')
    plt.legend(loc='best')
    
    #ADFuller test
    test_res = adfuller(ts, autolag = 'AIC')
    test_out = pd.Series(test_res[0:4], index = ['Test Statistics', 'P-value', '#Lags', 'Number of observations'] )
    for key, value in test_res[4].items():
        test_out['Critical Value (%s)'%key] = value
    print(test_out)
```

```python
test_stationarity(train['co2'])
```

```python
#removing Trend
moving_avg = train_log.rolling(52).mean()
stationarity_t = train_log - moving_avg
stationarity_t.dropna(inplace = True)

test_stationarity(stationarity_t)
```

```python
#making mean time independent
train_log_diff = train_log - train_log.shift(1)
test_stationarity(train_log_diff.dropna())
```

```python
#Removing Seasonal Fluctuations
decomp = seasonal_decompose(train_log, freq = 52)
trend = decomp.trend
seasonal = decomp.seasonal
residual = decomp.resid

plt.figure(figsize = (15, 6))
plt.subplot(411)
plt.plot(train_log, label = 'Original')
plt.legend(loc='best')

plt.subplot(412)
plt.plot(trend, label = 'Trend')
plt.legend(loc='best')

plt.subplot(413)
plt.plot(seasonal, label = 'Seasonal')
plt.legend(loc='best')


plt.subplot(414)
plt.plot(residual, label = 'Residual')
plt.legend(loc='best')

plt.tight_layout()
plt.show()
```

```python
#Checking stationarity of residuals
t_log_decomp = pd.DataFrame(residual)
t_log_decomp['Date'] = train_log.index
t_log_decomp.dropna(inplace = True)
test_stationarity(t_log_decomp['resid'])
```

```python
#Forecasting using Arima model
from statsmodels.tsa.stattools import acf, pacf
lag_acf = acf(train_log_diff.dropna(), nlags = 53)
lag_pacf = pacf(train_log_diff.dropna(), nlags = 53)
```

```python
plt.figure(figsize = (15, 8))
plt.plot(lag_acf)
plt.axhline(y = 0, linestyle = '--', color= 'grey')
plt.axhline(y = +1.96/np.sqrt(len(train_log_diff)), linestyle = '--', color = 'red')
plt.axhline(y = -1.96/np.sqrt(len(train_log_diff)), linestyle = '--', color = 'red')
plt.title('Auto correlation plot')
```

```python
plt.figure(figsize = (15, 8))
plt.plot(lag_pacf)
plt.axhline(y = 0, linestyle = '--', color= 'grey')
plt.axhline(y = +1.96/np.sqrt(len(train_log_diff)), linestyle = '--', color = 'red')
plt.axhline(y = -1.96/np.sqrt(len(train_log_diff)), linestyle = '--', color = 'red')
plt.title('Partial Auto correlation plot')
```

```python
#Prediction using ARIMA model
model = ARIMA(train_log, order=(10, 2, 6))
model
results_arima = model.fit(dist = -1)
plt.plot(train_log_diff.dropna(), label = 'original')
plt.plot(results_arima.fittedvalues, label = 'Predictions')
plt.legend(loc='best')
plt.plot()
```

### SARIMAX

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
fit_forecast = SARIMAX(train['co2'], order = (10,2,6), seasonal_order=(0,1,1,52)).fit()
```

```python
y_pred['SARIMAX'] = fit_forecast.predict(start='1998-01-01', end='2001-12-29')
plt.figure(figsize = (16, 8))
plt.plot(train['co2'], label = 'Train')
plt.plot(test['co2'], label = 'test')
plt.plot(y_pred['SARIMAX'], label = 'SARIMAX')
plt.legend(loc ='best')
```
